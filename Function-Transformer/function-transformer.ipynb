```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

# To plot the QQ plot 
import scipy.stats as stats 

from sklearn.preprocessing import FunctionTransformer,PowerTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
charges = df["charges"]
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(charges, kde=True)
plt.title('Original charges histogram')

plt.subplot(1,2,2)
stats.probplot(charges, dist="norm", plot=plt)
plt.title('Original charges QQ plot')

plt.tight_layout()
plt.show()
```

# Log transformation

```python
charges_log = np.log1p(charges)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(charges_log, kde=True)
plt.title('Log transformed charges')

plt.subplot(1,2,2)
stats.probplot(charges_log, dist="norm", plot=plt)
plt.title('Log QQ Plot')

plt.show()
```

# Square Root Transformation

```python
charges_sqrt = np.sqrt(charges)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(charges_sqrt, kde=True)
plt.title('Sqrt Transformed Charges')

plt.subplot(1,2,2)
stats.probplot(charges_sqrt, dist="norm", plot=plt)
plt.title('Sqrt QQ Plot')

plt.show()
```

# Reciprocal Transformation

```python
charges_recip = 1/charges
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(charges_recip, kde=True)
plt.title('Reciprocal Transformed Charges')

plt.subplot(1,2,2)
stats.probplot(charges_recip, dist="norm", plot=plt)
plt.title('Reciprocal QQ Plot')

plt.show()
```

# Square Transformation

```python
charges_square = charges ** 2
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(charges_square, kde=True)
plt.title('Square Transformed Charges')

plt.subplot(1,2,2)
stats.probplot(charges_square, dist="norm", plot=plt)
plt.title('Square QQ Plot')

plt.show()
```
